```python
# Definir a pasta em que os arquivos estão e onde irão
origem = '/content/drive/My Drive/Data1/COMEXSTAT1/'
```

```python
# Definir o nome dos arquivos
arquivo_1 = origem + 'EXP_2024_JJ.csv'
arquivo_2 = origem + 'EXP_2024_JD.csv'
arquivo_3 = origem + 'IMP_2024_JJ.csv'
arquivo_4 = origem + 'IMP_2024_JD.csv'
arquivo_5 = origem + 'EXP_2025_JJ.csv'
arquivo_6 = origem + 'EXP_2025_JD.csv'
arquivo_7 = origem + 'IMP_2025_JJ.csv'
arquivo_8 = origem + 'IMP_2025_JD.csv'
```

```python
# Fazer a leitura dos arquivos .csv
exp24JJ = pd.read_csv(arquivo_1,low_memory=False,sep =';',encoding='UTF-8')
exp24JD = pd.read_csv(arquivo_2,low_memory=False,sep =';',encoding='UTF-8')
imp24JJ = pd.read_csv(arquivo_3,low_memory=False,sep =';',encoding='UTF-8')
imp24JD = pd.read_csv(arquivo_4,low_memory=False,sep =';',encoding='UTF-8')
exp25JJ = pd.read_csv(arquivo_5,low_memory=False,sep =';',encoding='UTF-8')
exp25JD = pd.read_csv(arquivo_6,low_memory=False,sep =';',encoding='UTF-8')
imp25JJ = pd.read_csv(arquivo_7,low_memory=False,sep =';',encoding='UTF-8')
imp25JD = pd.read_csv(arquivo_8,low_memory=False,sep =';',encoding='UTF-8')
```

```python
# Concatenar os arquivos e separá-los em duas tabelas (Exp/Imp)
EXP_2024_25 = pd.concat([exp24JJ,exp24JD,exp25JJ,exp25JD], ignore_index=True)
IMP_2024_25 = pd.concat([imp24JJ,imp24JD,imp25JJ,imp25JD], ignore_index=True)
```

```python
# Fazer a limpeza dos dados e remover todos os valores nulos
print('Contagem de valores 0 por coluna (EXP_2024_25):')
zeros_count_exp = (EXP_2024_25 == 0).sum()
display(zeros_count_exp)

initial_rows_exp = EXP_2024_25.shape[0]
exp_final_cleaned = EXP_2024_25[~(EXP_2024_25 == 0).any(axis=1)]
rows_removed_exp = initial_rows_exp - exp_final_cleaned.shape[0]

print(f"\nNúmero original de linhas (EXP_2024_25): {initial_rows_exp}")
print(f"Número de linhas após remover '0's (EXP_2024_25): {exp_final_cleaned.shape[0]}")
print(f"Número de linhas removidas (EXP_2024_25): {rows_removed_exp}")

print("\nPrimeiras 5 linhas do DataFrame EXP_2024_25 limpo:")
display(exp_final_cleaned.head())

print('\n' + '='*50 + '\n') # Separator for clarity

print('Contagem de valores 0 por coluna (IMP_2024_25):')
zeros_count_imp = (IMP_2024_25 == 0).sum()
display(zeros_count_imp)

initial_rows_imp = IMP_2024_25.shape[0]
imp_final_cleaned = IMP_2024_25[~(IMP_2024_25 == 0).any(axis=1)]
rows_removed_imp = initial_rows_imp - imp_final_cleaned.shape[0]

print(f"\nNúmero original de linhas (IMP_2024_25): {initial_rows_imp}")
print(f"Número de linhas após remover '0's (IMP_2024_25): {imp_final_cleaned.shape[0]}")
print(f"Número de linhas removidas (IMP_2024_25): {rows_removed_imp}")

print("\nPrimeiras 5 linhas do DataFrame IMP_2024_25 limpo:")
display(imp_final_cleaned.head())
```

```python
# Dizer quantas linhas com valor "0" foram removidas
total_rows_removed = rows_removed_exp + rows_removed_imp
print(f"\nTotal de linhas removidas de ambos os DataFrames: {total_rows_removed}")
```

```python
# Exportar os arquivos finais para a pasta no DRIVE
exp_final_cleaned.to_csv(origem + 'EXP_FINAL.csv', index=False)
imp_final_cleaned.to_csv(origem + 'IMP_FINAL.csv', index=False)
```
